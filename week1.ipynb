## Logistic Regression

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# 1. We create our historical "Training Data" (The Blueprint)
# Features (X): Age, Income, Education
# Label (Y): Bought_Car (1 = Yes, 0 = No)
data = {
    'Age': [25, 45, 35, 50, 23],
    'Income': [30000, 80000, 45000, 100000, 25000],
    'Education': ["High School", "College", "College", "College", "High School"],
    'Bought_Car': [0, 1, 0, 1, 0]
}

# Convert the dictionary into a Pandas DataFrame (a spreadsheet)
df = pd.DataFrame(data)
print("--- Historical Data ---")
print(df)
```

```python
# Separate our Ingredients (X) and our Final Answer (Y)
# INTENTIONAL BUG: If we include 'Education' as text, the math fails!
X_buggy = df[['Age', 'Income', 'Education']]
Y = df['Bought_Car']
model = LogisticRegression()
model.fit(X_buggy, Y)

```

```python
# The Fix: Replace text with numbers
df['Education'] = df['Education'].map({'High School': 1, 'College': 2})

# Now we separate our clean numerical Features (X) and Labels (Y)
X = df[['Age', 'Income', 'Education']]
Y = df['Bought_Car']

# Create the "Brain" (The Logistic Regression Model)
model = LogisticRegression()

# Train the model (This is where the computer automatically adjusts the weights!)
print("\nTraining the model...")
model.fit(X, Y)
print("Model trained successfully!")
```

```python
model.coef_
```

```python
# Define the new customer data (Must match the exact structure of X)
new_customer = pd.DataFrame({
    'Age': [40],
    'Income': [70000],
    'Education': [2]
})

# Ask the AI to make a decision
prediction = model.predict(new_customer)
probability = model.predict_proba(new_customer)

print("\n--- AI Prediction ---")
print(f"Will the new customer buy a car? Predict: {prediction[0]} (1=Yes, 0=No)")
print(f"Confidence of buying: {probability[0][1]:.2%}")
```

## Neural Network (NN)

```python
X_buggy = df[['Age', 'Income', 'Education']]
Y = df['Bought_Car']
```

```python
from sklearn.neural_network import MLPClassifier
# Create a Neural Network
# hidden_layer_sizes=(4, 2) means:
# Layer 1 has 4 neurons, Layer 2 has 2 neurons.
neural_net_model = MLPClassifier(
    hidden_layer_sizes=(4, 2),
    max_iter=1000,
    random_state=42
)

# Train the Neural Network
print("Training the Neural Network...")
neural_net_model.fit(X_buggy, Y)
print("Neural Network trained successfully!")


```

```python
# Make a prediction using our stacked layers
nn_pred = neural_net_model.predict(new_customer)
nn_prob = neural_net_model.predict_proba(new_customer)

print("\n--- NEURAL NETWORK PREDICTION ---")
print(f"Prediction: {nn_pred[0]} (1=Yes, 0=No)")
print(f"Confidence of buying: {nn_prob[0][1]:.2%}")
```
